```python
import time
from openai import OpenAI
from dotenv import load_dotenv
import os
import random
import string
load_dotenv()  # Load .env file
oai_key = os.environ.get('oai')

def generate_random_text(length):
    return ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(length))

toshbox = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
max_tokens = 500
iterations = 100

start = time.time()
for i in range(iterations):
    response = toshbox.chat.completions.create(
        messages=[
            {
                'role': 'user',
                'content': generate_random_text(10),
            }
        ],
        model='openchat',
        # stream=True,
        max_tokens=max_tokens
    )
end = time.time()
print("toshbox: " + str((end-start)) + " seconds")
print("toshbox: " + str((iterations*max_tokens)/(end-start)) + " tok/s")
```
